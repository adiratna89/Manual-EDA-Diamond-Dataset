```python
numeric_df = df.select_dtypes(include = 'number')
print(numeric_df)
```

```python
categoric_df = [ i for i in df.columns if df[i].dtypes == 'category']
```

```python
categoric_df
```

```python
corr = numeric_df.corr() # finding co-relation
print(corr)
```

```python
plt.figure(figsize = (9,6))
sns.heatmap(corr, annot = True, cmap = 'coolwarm',cbar = True )
plt.title('Correlation Heatmap')
plt.show()
```
